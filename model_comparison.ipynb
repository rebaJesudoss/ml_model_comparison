##Checking normality of data using Kolmogorov-Smirnov Test

```python
from scipy.stats import kstest
#kstest(data, 'norm')

for i in range(2,11): 
  print (data.iloc[i]) 
  a,b= kstest(data.iloc[[i]], 'norm') 
  print ("Statistics", a, "p-value", b)
  if b < 0.05:
    print("The null hypothesis can be rejected") 
  else: 
    print("The null hypothesis cannot be rejected")
```

##Data is not normally distributed. So we nornalise the data



```python
trans_date=[]
for i in range(0,414):
  trans_date.append(data.iloc[i]["Transaction date"])
```

```python
data=data.drop(columns="Transaction date")
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
names = data.columns
names=list(names)

d = scaler.fit_transform(data)
scaled_data = pd.DataFrame(d, columns=names)
scaled_data.head()
```

```python
#trans_datee=pd.DataFrame({'Transaction date':trans_date})
scaled_data.insert(0, "Transaction date", trans_date, True)
scaled_data['Total price']=scaled_data["House size (sqft)"] * scaled_data["House price of unit area"]
scaled_data
```

##Splitting data into train and test

```python
from sklearn.model_selection import train_test_split

X = scaled_data[['House Age', 'Distance from nearest Metro station (km)', 'Number of convenience stores', 'latitude','longitude','Number of bedrooms','House size (sqft)']].values
y = scaled_data['House price of unit area'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 100)
```

##Using 3 models and finding the MAE

```python
#Decision Tree 

decision_model = DecisionTreeRegressor()  
decision_model.fit(X_train, y_train) 
predicted_decision_trees = decision_model.predict(X_test)
print ("Mean Absolute Error using Decision Tress :", mean_absolute_error(y_test, predicted_decision_trees))
```

```python
#Random Forest  

forest_model = RandomForestRegressor(n_estimators=100, max_depth=20)
forest_model.fit(X_train, y_train )
predicted_random_forest = forest_model.predict(X_test)
print("Mean Absolute Error using Random Forest:", mean_absolute_error(y_test, predicted_random_forest))
```

```python
#Multiple Linear 
reg = LinearRegression()
reg.fit(X_train, y_train)
reg_predict = reg.predict(X_test)
print("Mean Absolute Error using Multiple regression:", mean_absolute_error(y_test, reg_predict))
```

##From the error values, we find that Random forest is the best algorithm for Housing price prediction using a normalised data
